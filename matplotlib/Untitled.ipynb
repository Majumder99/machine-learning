```python
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
x = [1,2,3,4]
y = [2,3,4,5]

plt.plot(x,y,'g')
plt.title("info")
plt.xlabel("x")
plt.ylabel("y")
```

```python
plt.bar(x,y,color='g')
plt.title("info")
plt.xlabel("x")
plt.ylabel("y")
```

```python
plt.hist(x,y,color='g')
plt.title("info")
plt.xlabel("x")
plt.ylabel("y")
```

```python
plt.scatter(x,y)
plt.title("info")
plt.xlabel("x")
plt.ylabel("y")
```
